```python
import pandas as pd
import time
from src.utils import run_evaluation, run_evaluation_convo
import json
```

```python
# load environment variable from .env file
from dotenv import load_dotenv
load_dotenv()
```

```python
# Simple Questions
data = pd.read_csv("benchmark_gpt/simple_q_mapped.csv")
with open("benchmark_gpt/simple_response.json", "r") as f:
    res = json.load(f)
data_res = pd.DataFrame.from_records(res["answers"])

data_all_simple = pd.merge(data, data_res, left_on="answerable_question", right_on="question").drop(columns=["question"])
data_all_simple.head()

# Medium Questions
data = pd.read_csv("benchmark_gpt/medium_q_mapped.csv")
with open("benchmark_gpt/medium_response.json", "r") as f:
    res = json.load(f)
data_res = pd.DataFrame.from_records(res["answers"])
data_all_medium = pd.merge(data, data_res, left_on="answerable_question", right_on="question").drop(columns=["question"])
data_all_medium.head()
# Hard Questions
data = pd.read_csv("benchmark_gpt/hard_q_mapped.csv")
with open("benchmark_gpt/hard_response.json", "r") as f:
    res = json.load(f)
data_res = pd.DataFrame.from_records(res["answers"])
data_all_hard = pd.merge(data, data_res, left_on="answerable_question",
    right_on="question").drop(columns=["question"])
data_all_hard.head()

```

```python
data_all_simple.shape, data_all_medium.shape, data_all_hard.shape
```

```python
# shape where answer is not null
data_all_simple[data_all_simple["answer"].notnull()].shape, data_all_medium[data_all_medium["answer"].notnull()].shape, data_all_hard[data_all_hard["answer"].notnull()].shape
```

```python
def simple():
    results = run_evaluation( dataframe=data_all_simple
                            , question_column="answerable_question"
                            , response_column="answer"
                            , eval_type="gpt_simple"
                            , output_dir="out_gpt", batch_size=2)

    results = run_evaluation_convo( dataframe=data_all_simple
                            , question_column="answerable_question"
                            , response_column="answer"
                            , eval_type="gpt_simple_convo"
                            , output_dir="out_gpt", batch_size=2)
    results.head()

def medium():
    results = run_evaluation( dataframe=data_all_medium
                            , question_column="answerable_question"
                            , response_column="answer"
                            , eval_type="gpt_medium"
                            , output_dir="out_gpt", batch_size=2)

    results = run_evaluation_convo( dataframe=data_all_medium
                            , question_column="answerable_question"
                            , response_column="answer"
                            , eval_type="gpt_medium_convo"
                            , output_dir="out_gpt", batch_size=2)
    results.head()
def hard():
    results = run_evaluation( dataframe=data_all_hard
                            , question_column="answerable_question"
                            , response_column="answer"
                            , eval_type="gpt_hard"
                            , output_dir="out_gpt", batch_size=2)

    results = run_evaluation_convo( dataframe=data_all_hard
                            , question_column="answerable_question"
                            , response_column="answer"
                            , eval_type="gpt_hard_convo"
                            , output_dir="out_gpt", batch_size=2)
    results.head()
```

## Simple

```python
simple()
medium()
hard()
```
