```python
import pandas as pd
```

```python
data_medium = pd.read_csv("benchmark_ppl/medium_q_mapped.csv")
data_hard = pd.read_csv("benchmark_ppl/hard_q_mapped.csv")

data_medium_org = pd.read_csv("benchmark_ppl/old/medium_q.csv")
data_hard_org = pd.read_csv("benchmark_ppl/old/hard_q.csv")
```

```python
data_medium.columns, data_medium_org.columns, data_medium_org.shape, data_medium.shape
```

```python
data_medium_org = data_medium_org.merge(data_medium, left_on='questions', right_on='original_question', how='left').drop(columns=["questions"]).reset_index(drop=True)  
data_medium_org.shape
```

```python
data_medium_org.head()
```

```python
data_medium_org.to_csv("benchmark_ppl/medium_q_mapped.csv", index=False)
```

```python
data_hard.columns, data_hard_org.columns, data_hard_org.shape, data_hard.shape
```

```python
data_hard_org = data_hard_org.merge(data_hard, left_on='synthetic_questions', right_on='original_question', how='left').drop(columns=["synthetic_questions", "Unnamed: 4"]).reset_index(drop=True)  
data_hard_org.shape
```

```python
data_hard_org.to_csv("benchmark_ppl/hard_q_mapped.csv", index=False)
```
